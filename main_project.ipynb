```python
# here we are sort the values of the superheros according to their speed.
# False means that the list will be sorted in ascending order. True means that the list will be sorted in descending (reverse) order

speed_data = good.sort_values(by=['Speed'],ascending=False).head()
speed_data
```

```python
max_speed_good = good[good['Speed'] == 100]
max_speed_good
```

```python
# getting the most powerful superhero by sortinf
good.sort_values(by=['Power'],ascending=False).head()
```

```python
good_power = good[good['Power'] == 100]
good_power.shape
```

```python
good_max_power = good.sort_values(by=['Total'],ascending=False)
good_max_power
```

```python
good_max_power.head()
```

```python
# We are visulizing the data for the good superheroes with max-power
plt.figure(figsize=(8,5))  
plt.bar(list(good_max_power['Name'])[0:5], list(good_max_power['Total'])[0:5],color='b')
plt.xlabel('Name')
plt.ylabel('Power')
plt.show()


# so here [0:5] means we got first five values.
# figsize=(8,5) width:8 and height:5
```
